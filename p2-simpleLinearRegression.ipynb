```python
# Simple linear Regression :

from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


# Loading the dataset
data = fetch_california_housing()

# Load only 1 column in x as it simple linear regression
x=data.data[:,[0]]
y=data.target


# Check data loadin once
print("X :",x)
print("Y :",y)

# Split dataset into training dataset and test dataset
X_train, X_test ,Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
# Evaluate

y_pred = lr.predict(X_test)
print("Mean squared error : ",mean_squared_error(Y_test,y_pred))
print("R2 score : ",r2_score(Y_test,y_pred))
```

```python
# Plotting
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_test.ravel(), y=Y_test, alpha=0.7, color='blue', label='Actual Data')
sns.regplot(x=X_test.ravel(), y=y_pred, scatter=False, color='red', line_kws={'linewidth': 2}, label='Regression Line')

plt.xlabel("Feature", fontsize=12)
plt.ylabel("Target", fontsize=12)
plt.title("Simple Linear Regression", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
